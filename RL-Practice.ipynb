```python
class KerasEnv:
    def __init__(self):
        self.env=tf.keras.Sequential()
        self.env.add(tf.keras.layers.Dense(state_dim+1,input_shape=(state_dim+1,),activation='sigmoid'))
        
    def get_observation(self,state,action):
        action=np.expand_dims(np.expand_dims(action, axis=0),axis=0)
        s=np.concatenate((state,action),axis=1)
        obs=self.env.predict(s)
        s=obs[0,0:obs.shape[1]-1].reshape(1,state_dim)
        r=obs[0,obs.shape[1]-1]
    
        return (s,r)
        
     
   
```

```python
class Grid:
    def __init__(self):
        self.grid=np.array([[0,10,0],[10,100,10],[100,1000,100]])
        self.punishment=-1000
        
        
    def get_observation(self,state,action):
        x=state%3
        y=(int)(state/3)
        next_state=np.zeros((no_of_actions)).astype('int')
        reward=np.zeros((no_of_actions))
        
        
  
        if(x>0):
            next_state[0]=(x-1)+y*3  
            reward[0]=self.grid[x-1][y]
        
        else:
            next_state[0]=state
            reward[0]=self.punishment
        
        if(x<2):
            next_state[1]=(x+1)+y*3 
            reward[1]=self.grid[x+1][y]
        
        else:
            next_state[1]=state
            reward[1]=self.punishment
        
        if(y>0):
            next_state[2]=(x)+(y-1)*3  
            reward[2]=self.grid[x][y-1]
        
        else:
            next_state[2]=state
            reward[2]=self.punishment
        
        if(y<2):
            next_state[3]=(x)+(y+1)*3  
            reward[3]=self.grid[x][y+1]
        
        else:
            next_state[3]=state
            reward[3]=self.punishment
        
        
            
        return (next_state[action],reward)
    
def get_action(model,state):
    a=np.array(np.random.choice([np.argmax(model[state]),np.random.randint(4)],p=[eps,1-eps]))
    return a
        
```

```python
class Grid:
    def __init__(self):
        self.grid=np.array([[0,3,0],[3,50,3],[50,1000,50]])
        
        
    def get_observation(self,st,action):
        state=[[np.argmax(st[0][0:3]),np.argmax(st[0][3:])]]
        next_state=np.zeros((1,state_dim))
        #print(state[0][0],(int)(state[0][1]))
#       print(self.grid[state[0][0]][state[0][1]+1])
        if (action==0 and state[0][1]>0):
            ind=[state[0][0],state[0][1]-1+3]
            next_state[0][ind]=1
            reward=self.grid[state[0][0]][state[0][1]-1]
        elif (action==1 and state[0][1]<2):
            ind=[state[0][0],state[0][1]+1+3]
            next_state[0][ind]=1
            reward=self.grid[state[0][0]][state[0][1]+1]
        elif (action==2 and state[0][0]>0):
            ind=[state[0][0]-1,state[0][1]+3]
            next_state[0][ind]=1
            reward=self.grid[state[0][0]-1][state[0][1]]
        elif (action==3 and state[0][0]<2):
            ind=[state[0][0]+1,state[0][1]+3]
            next_state[0][ind]=1
            reward=self.grid[state[0][0]+1][state[0][1]]        
        else: 
            next_state=st
            reward=-1000
            
        return (next_state,reward)
        
```

```python
def getModel():
    model=tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(3,input_shape=(state_dim,),activation='sigmoid'))
    model.add(tf.keras.layers.Dense(no_of_actions,activation='relu'))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),loss='mse')
    return model

def get_action(state):
    actions=model.predict(state)
    #print(actions)
    a=np.array(np.random.choice([np.argmax(actions,1)[0],np.random.randint(3)],p=[eps,1-eps]))
    return a
```

```python
model=getModel()
target_model=getModel()
target_model.set_weights(model.get_weights()) 
```
