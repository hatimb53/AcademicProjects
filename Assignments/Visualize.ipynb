```python
import matplotlib.pyplot as plt
import numpy as np
x=[]
y=[]
a=1
b=1
for i in range(100):
    
    for j in range(100):
        x.append(i)
        y.append(j)
x=np.array(x)
y=np.array(y)
Q=x+y

C=a*x**2+b*(Q-x)**2


from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#ax.scatter(x,y,Q)
ax.scatter(x,Q-x,C)


ax.set_ylim(0,100)
ax.set_xlim(0,100)

plt.show()
```
