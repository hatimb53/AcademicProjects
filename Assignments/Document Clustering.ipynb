```python
import urllib3
http = urllib3.PoolManager()
from sklearn.feature_extraction.text import TfidfVectorizer
transformer=TfidfVectorizer(smooth_idf=False)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0)
```

```python
r = http.request('GET','https://www.cs.umb.edu/~smimarog/textmining/datasets/r52-train-all-terms.txt')
```

```python
import numpy as np
data=r.data.splitlines()
def transform():
    li=str(x).split('\\t')
    li[0]
data=np.array(list(map(lambda x: np.array(str(x).split('\\t')),data)))
data[:,]
```

```python
data=r.data.splitlines()
data=transformer.fit_transform(data)
```

```python
kmeans.fit(data)
```

```python
kmeans.labels_
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=3, random_state=42)
data1 = svd.fit_transform(data) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(data1[:,0],data1[:,1],c=kmeans.labels_)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data1[:,0],data1[:,1],data1[:,2], c=kmeans.labels_ )
plt.show()
```
