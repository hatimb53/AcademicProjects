```python
def rotation(theta):
    return np.array([[math.cos(theta*math.pi/180),-math.sin(theta*math.pi/180)],[math.sin(theta*math.pi/180),math.cos(theta*math.pi/180)]])

def scale(k,m):
    return np.array([[k,0],[0,m]])
def shear():
    return np.array([[1,1],[0,1]])
```
