```python

def textjustify(s,wd):   
    word=s.split(" ")
    line=[]
    width=wd
    
    flag=True
    s=''
    count=0

    while count<len(word):

        #print(word[count]," ",width)
        if(width>=len(word[count])+1 or (flag==True and width>=len(word[count]))):
            #print(word[count])
            if flag==True:
                s=word[count]
                
                flag=False
            else:
                s=s+' '+word[count]
                width=width-1

            width=width-len(word[count])
            count=count+1

            #print(width)
        else:
            line.append(s)
            width=wd
            flag=True
            


    line.append(s)
    return line
```
